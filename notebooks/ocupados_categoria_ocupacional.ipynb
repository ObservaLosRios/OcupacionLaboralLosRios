````python
# â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬
# ANÃLISIS DE TRABAJOS MÃS COMUNES - DISEÃ‘O EJECUTIVO THE ECONOMIST
# â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬

# Analizar los trabajos mÃ¡s comunes (excluyendo el total)
trabajos_principales = datos[
    (datos['Grupo_Simple'] != 'TOTAL OCUPADOS') & 
    (datos['Sexo'] == 'Ambos sexos')
].copy()

# Calcular el total de ocupados por tipo de trabajo y eliminar duplicados de manera robusta
ranking_trabajos = trabajos_principales.groupby('Grupo_Simple')['Value'].sum().reset_index()
ranking_trabajos = ranking_trabajos.sort_values('Value', ascending=True)

# Eliminar duplicados de manera mÃ¡s robusta
print("ğŸ” DEPURACIÃ“N: Sectores antes de eliminar duplicados:")
print(f"   Total de registros: {len(ranking_trabajos)}")
for i, trabajo in enumerate(ranking_trabajos['Grupo_Simple'].values):
    print(f"   {i+1}. {trabajo}")

# Eliminar duplicados manteniendo el primer registro
ranking_trabajos_unicos = ranking_trabajos.drop_duplicates(subset=['Grupo_Simple'], keep='first').reset_index(drop=True)

print(f"\nâœ… RESULTADO: Sectores Ãºnicos despuÃ©s de limpieza:")
print(f"   Total de registros Ãºnicos: {len(ranking_trabajos_unicos)}")
for i, trabajo in enumerate(ranking_trabajos_unicos['Grupo_Simple'].values):
    print(f"   {i+1}. {trabajo}")

# Verificar que no hay duplicados
grupos_unicos = ranking_trabajos_unicos['Grupo_Simple'].nunique()
total_registros = len(ranking_trabajos_unicos)
print(f"\nğŸ” VERIFICACIÃ“N: {grupos_unicos} grupos Ãºnicos de {total_registros} registros totales")

# Usar los datos Ãºnicos para el resto del anÃ¡lisis
ranking_trabajos = ranking_trabajos_unicos.copy()

# â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬
# PALETA VISUAL PREMIUM - ESTILO THE ECONOMIST
# â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬

# Paleta de colores The Economist refinada y elegante
colores_the_economist_premium = {
    'navy_deep': '#1e3a8a',           # Azul marino profundo
    'burgundy_elegant': '#7f1d1d',    # Burdeos elegante  
    'forest_sophisticated': '#14532d', # Verde bosque sofisticado
    'amber_executive': '#d97706',      # Ãmbar ejecutivo
    'slate_premium': '#334155',        # Pizarra premium
    'crimson_refined': '#dc2626',      # CarmesÃ­ refinado
    'emerald_corporate': '#047857',    # Esmeralda corporativo
    'gold_luxury': '#ca8a04',          # Oro de lujo
    'teal_sophisticated': '#0f766e',   # Verde azulado sofisticado
    'violet_executive': '#7c3aed'      # Violeta ejecutivo
}

# Crear etiquetas mejoradas con algoritmo inteligente de saltos de lÃ­nea
etiquetas_trabajos = []
for trabajo in ranking_trabajos['Grupo_Simple']:
    # Algoritmo mejorado para manejar nombres largos
    if len(trabajo) <= 20:
        # Nombres cortos: mantener en una lÃ­nea
        etiquetas_trabajos.append(trabajo)
    elif ' y ' in trabajo:
        # Casos especiales como "Directivos y Gerentes"
        partes = trabajo.split(' y ')
        if len(partes) == 2:
            etiquetas_trabajos.append(f"{partes[0]}<br>y {partes[1]}")
        else:
            etiquetas_trabajos.append(trabajo)
    elif len(trabajo.split()) >= 3:
        # Nombres con 3 o mÃ¡s palabras: buscar punto de corte Ã³ptimo
        palabras = trabajo.split()
        mitad = len(palabras) // 2
        etiquetas_trabajos.append(f"{' '.join(palabras[:mitad])}<br>{' '.join(palabras[mitad:])}")
    else:
        # Nombres de 2 palabras: dividir en el espacio
        palabras = trabajo.split()
        etiquetas_trabajos.append(f"{palabras[0]}<br>{palabras[1]}")

print(f"\nğŸ·ï¸ Etiquetas creadas: {len(etiquetas_trabajos)}")
print("ğŸ“ Etiquetas con formato:")
for i, etiqueta in enumerate(etiquetas_trabajos):
    print(f"   {i+1}. {etiqueta.replace('<br>', ' â†’ ')}")

# Aplicar colores Ãºnicos y profesionales The Economist
colores_aplicados_trabajos = list(colores_the_economist_premium.values())[:len(ranking_trabajos)]


# Crear grÃ¡fico ejecutivo profesional
fig = go.Figure()

# Calcular porcentajes para las anotaciones
total_empleos = ranking_trabajos['Value'].sum()
ranking_trabajos['Porcentaje'] = (ranking_trabajos['Value'] / total_empleos) * 100

# Crear grÃ¡fico de barras horizontales con diseÃ±o The Economist refinado
fig.add_trace(go.Bar(
    x=ranking_trabajos['Value'],
    y=etiquetas_trabajos,
    orientation='h',
    marker=dict(
        color=colores_aplicados_trabajos,
        line=dict(color='white', width=1.5),
        opacity=0.85
    ),
    hovertemplate='<b>%{y}</b><br>' +
                  'Total empleados: %{x:,.0f}<br>' +
                  'ParticipaciÃ³n: %{customdata:.1f}% del empleo total<br>' +
                  '<extra></extra>',
    customdata=ranking_trabajos['Porcentaje'],
    showlegend=False
))

# Agregar anotaciones de porcentaje al lado derecho con estilo The Economist
for i, (valor, porcentaje) in enumerate(zip(ranking_trabajos['Value'], ranking_trabajos['Porcentaje'])):
    fig.add_annotation(
        x=valor + (ranking_trabajos['Value'].max() * 0.03),
        y=i,
        text=f'<b>{porcentaje:.1f}%</b>',
        showarrow=False,
        font=dict(size=11, color='#1e293b', family='Georgia, serif'),
        align='left',
        xanchor='left',
        yanchor='middle'
    )

# LÃ­nea de promedio con estilo The Economist elegante
promedio_empleos = ranking_trabajos['Value'].mean()
fig.add_vline(
    x=promedio_empleos,
    line=dict(dash="dot", color='#64748b', width=2),
    annotation=dict(
        text=f"<b>Promedio<br>{promedio_empleos:,.0f}</b>", 
        textangle=0,
        font=dict(size=11, color='#1e293b', family='Georgia, serif'),
        bgcolor='rgba(255,255,255,0.95)',
        bordercolor='#64748b',
        borderwidth=1
    ),
    annotation_position="top"
)

# â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬
# DISEÃ‘O EJECUTIVO PREMIUM - ESTILO THE ECONOMIST
# â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬

fig.update_layout(
    title=dict(
        text='<b>Los Trabajos MÃ¡s Comunes en la RegiÃ³n de Los RÃ­os</b><br>' +
             '<span style="font-size:14px; color:#64748b;">â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬</span><br>' +
             '<span style="font-size:13px; color:#475569;">DistribuciÃ³n del empleo por sector ocupacional â€¢ Los RÃ­os, Chile (2010-2019)</span>',
        font=dict(size=18, color='#1e293b', family='Georgia, serif'),
        x=0.12,  # Mover tÃ­tulo mÃ¡s a la derecha
        xanchor='left',
        pad=dict(b=30)
    ),
    xaxis=dict(
        title=dict(
            text='<b>NÃºmero Total de Personas Empleadas</b>',
            font=dict(size=13, color='#334155', family='Georgia, serif')
        ),
        tickformat=',.0f',
        tickfont=dict(size=11, color='#475569', family='Georgia, serif'),
        gridcolor='#e2e8f0',
        gridwidth=1,
        showgrid=True,
        zeroline=False,
        linecolor='#cbd5e1',
        linewidth=1
    ),
    yaxis=dict(
        title='',
        tickfont=dict(size=10, color='#1e293b', family='Georgia, serif'),
        automargin=True,
        tickmode='linear',
        categoryorder='array',
        categoryarray=etiquetas_trabajos,
        linecolor='#cbd5e1',
        linewidth=1,
        showgrid=False
    ),
    plot_bgcolor='#fefefe',
    paper_bgcolor='white',
    height=750,
    showlegend=False,
    margin=dict(l=280, r=140, t=160, b=90),  # Aumentar margen izquierdo para mover grÃ¡fico a la derecha
    font=dict(family='Georgia, serif', color='#1e293b')
)

fig.show()

# ANÃLISIS EJECUTIVO PROFESIONAL DE TRABAJOS MÃS COMUNES
print("\n" + "="*75)
print("ğŸ¯ ANÃLISIS EJECUTIVO: ESTRUCTURA DEL EMPLEO REGIONAL")
print("="*75)

print("\nğŸ“‹ METODOLOGÃA Y CRITERIOS:")
print("â€¢ AnÃ¡lisis basado en nÃºmero absoluto de empleados por sector")
print("â€¢ Datos agregados del perÃ­odo 2010-2019 para estabilidad estadÃ­stica")
print("â€¢ EliminaciÃ³n robusta de duplicados para evitar doble conteo")
print("â€¢ ClasificaciÃ³n por importancia estratÃ©gica segÃºn participaciÃ³n")
print("â€¢ IdentificaciÃ³n de sectores motores de la economÃ­a regional")

# Mostrar estadÃ­sticas clave con interpretaciÃ³n ejecutiva
print(f"\nğŸ“Š INDICADORES ESTRATÃ‰GICOS:")
print("-" * 50)
print(f"ğŸ‘¥ Total de personas empleadas: {total_empleos:,.0f}")
print(f"ğŸ“ˆ Promedio por sector: {promedio_empleos:,.0f} empleados")
print(f"ğŸ¢ Sectores identificados: {len(ranking_trabajos)}")

# Top 3 trabajos mÃ¡s comunes con anÃ¡lisis detallado
top_3 = ranking_trabajos.tail(3)
print(f"\nğŸ† TOP 3 SECTORES ESTRATÃ‰GICOS:")
print("-" * 40)

for i, (idx, fila) in enumerate(top_3.iterrows(), 1):
    medalla = "ğŸ¥‡" if i == 3 else "ğŸ¥ˆ" if i == 2 else "ğŸ¥‰"
    
    # AnÃ¡lisis de concentraciÃ³n
    if fila['Porcentaje'] >= 25:
        concentracion = "CONCENTRACIÃ“N ALTA"
    elif fila['Porcentaje'] >= 15:
        concentracion = "CONCENTRACIÃ“N MODERADA"
    else:
        concentracion = "CONCENTRACIÃ“N BAJA"
    
    print(f"{medalla} {fila['Grupo_Simple']}")
    print(f"   â€¢ Empleados: {fila['Value']:,.0f} personas")
    print(f"   â€¢ ParticipaciÃ³n: {fila['Porcentaje']:.1f}% del empleo total")
    print(f"   â€¢ EvaluaciÃ³n: {concentracion}")

# ConcentraciÃ³n del empleo - anÃ¡lisis estratÃ©gico
concentracion_top3 = (top_3['Value'].sum() / total_empleos) * 100
concentracion_top5 = (ranking_trabajos.tail(5)['Value'].sum() / total_empleos) * 100

print(f"\nğŸ” ANÃLISIS DE CONCENTRACIÃ“N:")
print("-" * 35)
print(f"ğŸ“Š Top 3 sectores: {concentracion_top3:.1f}% del empleo total")
print(f"ğŸ“Š Top 5 sectores: {concentracion_top5:.1f}% del empleo total")

# InterpretaciÃ³n estratÃ©gica de concentraciÃ³n
if concentracion_top3 > 60:
    nivel_concentracion = "ALTA - EconomÃ­a concentrada en pocos sectores"
    riesgo_estrategico = "ALTO - Vulnerabilidad ante cambios sectoriales"
    recomendacion_concentracion = "Diversificar la base econÃ³mica regional"
elif concentracion_top3 > 45:
    nivel_concentracion = "MODERADA - Equilibrio sectorial aceptable"
    riesgo_estrategico = "MEDIO - Estabilidad con oportunidades de mejora"
    recomendacion_concentracion = "Fortalecer sectores emergentes"
else:
    nivel_concentracion = "BAJA - EconomÃ­a diversificada"
    riesgo_estrategico = "BAJO - Resiliencia ante cambios"
    recomendacion_concentracion = "Mantener diversificaciÃ³n actual"

print(f"\nğŸ’¡ INTERPRETACIÃ“N ESTRATÃ‰GICA:")
print("-" * 40)
print(f"â€¢ Nivel de concentraciÃ³n: {nivel_concentracion}")
print(f"â€¢ Riesgo estratÃ©gico: {riesgo_estrategico}")
print(f"â€¢ RecomendaciÃ³n: {recomendacion_concentracion}")

# Oportunidades y recomendaciones ejecutivas
print(f"\nğŸš€ OPORTUNIDADES DE DESARROLLO:")
print("-" * 45)
print("â€¢ Fortalecer sectores con alto potencial de crecimiento")
print("â€¢ Desarrollar programas de capacitaciÃ³n laboral especÃ­ficos")
print("â€¢ Promover inversiÃ³n en sectores estratÃ©gicos identificados")
print("â€¢ Crear polÃ­ticas de apoyo a sectores complementarios")
print("â€¢ Impulsar la innovaciÃ³n en sectores tradicionales")

print(f"\nğŸ“ˆ RECOMENDACIONES PARA TOMADORES DE DECISIÃ“N:")
print("-" * 55)
print("â€¢ Implementar planes de desarrollo sectorial diferenciados")
print("â€¢ Establecer mesas de trabajo pÃºblico-privadas por sector")
print("â€¢ Crear incentivos para diversificaciÃ³n econÃ³mica")
print("â€¢ Desarrollar infraestructura de apoyo a sectores clave")
print("â€¢ Monitorear indicadores de empleo sectorial regularmente")

# â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬
# ANÃLISIS DE PARTICIPACIÃ“N DE GÃ‰NERO - DISEÃ‘O EJECUTIVO THE ECONOMIST
# â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬

print("ğŸ¯ ANÃLISIS DE PARTICIPACIÃ“N DE GÃ‰NERO")
print("="*60)
print("ğŸ“Š Creando visualizaciÃ³n ejecutiva estilo The Economist...")

# Analizar diferencias por gÃ©nero con datos limpios
datos_genero = datos[
    (datos['Grupo_Simple'] != 'TOTAL OCUPADOS') & 
    (datos['Sexo'].isin(['Hombres', 'Mujeres']))
].copy()

# Calcular totales por trabajo y sexo
genero_trabajo = datos_genero.groupby(['Grupo_Simple', 'Sexo'])['Value'].sum().reset_index()
genero_pivot = genero_trabajo.pivot(index='Grupo_Simple', columns='Sexo', values='Value').fillna(0)

# CALCULAR TODAS LAS MÃ‰TRICAS NECESARIAS PRIMERO
genero_pivot['Total'] = genero_pivot['Hombres'] + genero_pivot['Mujeres']
genero_pivot['Participacion_Mujeres'] = (genero_pivot['Mujeres'] / genero_pivot['Total']) * 100
genero_pivot['Participacion_Hombres'] = (genero_pivot['Hombres'] / genero_pivot['Total']) * 100

# Filtrar trabajos con datos suficientes y ordenar por participaciÃ³n femenina
genero_pivot = genero_pivot[genero_pivot['Total'] > 500].copy()
genero_pivot = genero_pivot.sort_values('Participacion_Mujeres', ascending=True)

print(f"âœ… Datos procesados: {len(genero_pivot)} sectores analizados")

# PALETA DE COLORES EJECUTIVA ESTILO THE ECONOMIST
colores_the_economist_genero = {
    'masculinizado_extremo': '#7f1d1d',    # Burdeos profundo
    'masculinizado': '#dc2626',            # Rojo clÃ¡sico
    'ligeramente_masculino': '#ea580c',    # Naranja ejecutivo
    'equilibrado': '#1e40af',              # Azul The Economist
    'ligeramente_femenino': '#0f766e',     # Verde azulado
    'feminizado': '#14532d',               # Verde profundo
    'feminizado_extremo': '#0f172a'        # Azul marino oscuro
}

# Crear etiquetas mejoradas SOLO para el eje Y
etiquetas_mejoradas = []
for trabajo in genero_pivot.index:
    if len(trabajo) > 25:
        palabras = trabajo.split(' ')
        if len(palabras) > 2:
            punto_corte = len(palabras) // 2
            linea1 = ' '.join(palabras[:punto_corte])
            linea2 = ' '.join(palabras[punto_corte:])
            etiquetas_mejoradas.append(f"{linea1}<br>{linea2}")
        else:
            etiquetas_mejoradas.append(trabajo)
    else:
        etiquetas_mejoradas.append(trabajo)

# Asignar colores segÃºn nivel de participaciÃ³n de gÃ©nero
colores_aplicados = []
for participacion in genero_pivot['Participacion_Mujeres']:
    if participacion < 15:
        colores_aplicados.append(colores_the_economist_genero['masculinizado_extremo'])
    elif participacion < 30:
        colores_aplicados.append(colores_the_economist_genero['masculinizado'])
    elif participacion < 45:
        colores_aplicados.append(colores_the_economist_genero['ligeramente_masculino'])
    elif participacion <= 55:
        colores_aplicados.append(colores_the_economist_genero['equilibrado'])
    elif participacion < 70:
        colores_aplicados.append(colores_the_economist_genero['ligeramente_femenino'])
    elif participacion < 85:
        colores_aplicados.append(colores_the_economist_genero['feminizado'])
    else:
        colores_aplicados.append(colores_the_economist_genero['feminizado_extremo'])

# CREAR GRÃFICO EJECUTIVO ESTILO THE ECONOMIST
fig = go.Figure()

fig.add_trace(go.Bar(
    x=genero_pivot['Participacion_Mujeres'],
    y=etiquetas_mejoradas,
    orientation='h',
    marker=dict(
        color=colores_aplicados,
        line=dict(color='white', width=1.2),
        opacity=0.85
    ),
    hovertemplate='<b>%{y}</b><br>' +
                  'ParticipaciÃ³n de mujeres: %{x:.1f}%<br>' +
                  'ParticipaciÃ³n de hombres: %{customdata[0]:.1f}%<br>' +
                  'Total de empleados: %{customdata[1]:,.0f}<br>' +
                  '<extra></extra>',
    customdata=np.column_stack([
        genero_pivot['Participacion_Hombres'],
        genero_pivot['Total']
    ]),
    showlegend=False
))

# LÃ­nea de referencia de paridad con estilo The Economist
fig.add_vline(
    x=50, 
    line=dict(dash="dot", color='#64748b', width=2),
    annotation=dict(
        text="<b>Paridad<br>50%</b>",
        textangle=0,
        font=dict(size=11, color='#1e293b', family='Georgia, serif'),
        bgcolor='rgba(255,255,255,0.95)',
        bordercolor='#64748b',
        borderwidth=1
    ),
    annotation_position="top"
)

# DISEÃ‘O EJECUTIVO PREMIUM ESTILO THE ECONOMIST
fig.update_layout(
    title=dict(
        text='<b>ParticipaciÃ³n de Mujeres por Sector Laboral</b><br>' +
             '<span style="font-size:14px; color:#64748b;">â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬</span><br>' +
             '<span style="font-size:13px; color:#475569;">DistribuciÃ³n de gÃ©nero en el empleo â€¢ Los RÃ­os, Chile (2010-2019)</span>',
        font=dict(size=18, color='#1e293b', family='Georgia, serif'),
        x=0.12,
        xanchor='left',
        pad=dict(b=30)
    ),
    xaxis=dict(
        title=dict(
            text='<b>ParticipaciÃ³n de Mujeres (%)</b>',
            font=dict(size=13, color='#334155', family='Georgia, serif')
        ),
        range=[0, 100],
        ticksuffix='%',
        tickfont=dict(size=11, color='#475569', family='Georgia, serif'),
        gridcolor='#e2e8f0',
        gridwidth=1,
        showgrid=True,
        zeroline=False,
        linecolor='#cbd5e1',
        linewidth=1
    ),
    yaxis=dict(
        title='',
        tickfont=dict(size=10, color='#1e293b', family='Georgia, serif'),
        automargin=True,
        showgrid=False,
        categoryorder='array',
        categoryarray=etiquetas_mejoradas,
        linecolor='#cbd5e1',
        linewidth=1
    ),
    plot_bgcolor='#fefefe',
    paper_bgcolor='white',
    height=650,
    showlegend=False,
    margin=dict(l=280, r=120, t=160, b=90),
    font=dict(family='Georgia, serif', color='#1e293b')
)

fig.show()

# ANÃLISIS EJECUTIVO PROFESIONAL
print("\n" + "="*75)
print("ğŸ‘¥ ANÃLISIS EJECUTIVO: EQUIDAD DE GÃ‰NERO EN EL EMPLEO")
print("="*75)

print("\nğŸ¯ METODOLOGÃA:")
print("â€¢ AnÃ¡lisis basado en distribuciÃ³n porcentual por sector")
print("â€¢ Filtro: Sectores con mÃ¡s de 500 empleados para relevancia estadÃ­stica") 
print("â€¢ ClasificaciÃ³n en 7 categorÃ­as de equidad de gÃ©nero")
print("â€¢ LÃ­nea de paridad al 50% como referencia visual")

# AnÃ¡lisis por categorÃ­as de gÃ©nero
masculinizados_extremo = genero_pivot[genero_pivot['Participacion_Mujeres'] < 15]
masculinizados = genero_pivot[(genero_pivot['Participacion_Mujeres'] >= 15) & (genero_pivot['Participacion_Mujeres'] < 30)]
equilibrados = genero_pivot[(genero_pivot['Participacion_Mujeres'] >= 45) & (genero_pivot['Participacion_Mujeres'] <= 55)]
feminizados = genero_pivot[(genero_pivot['Participacion_Mujeres'] > 70) & (genero_pivot['Participacion_Mujeres'] < 85)]
feminizados_extremo = genero_pivot[genero_pivot['Participacion_Mujeres'] >= 85]

print("\nğŸ” DISTRIBUCIÃ“N POR CATEGORÃAS:")
print("-" * 50)

if len(masculinizados_extremo) > 0:
    print(f"ğŸ”´ SECTORES ALTAMENTE MASCULINIZADOS (<15% mujeres): {len(masculinizados_extremo)}")
    for trabajo, datos_trab in masculinizados_extremo.iterrows():
        print(f"   â€¢ {trabajo}: {datos_trab['Participacion_Mujeres']:.1f}% mujeres | {datos_trab['Total']:,.0f} empleados")

if len(masculinizados) > 0:
    print(f"\nğŸŸ  SECTORES MASCULINIZADOS (15-30% mujeres): {len(masculinizados)}")
    for trabajo, datos_trab in masculinizados.iterrows():
        print(f"   â€¢ {trabajo}: {datos_trab['Participacion_Mujeres']:.1f}% mujeres | {datos_trab['Total']:,.0f} empleados")

if len(equilibrados) > 0:
    print(f"\nğŸŸ¡ SECTORES EQUILIBRADOS (45-55% mujeres): {len(equilibrados)}")
    for trabajo, datos_trab in equilibrados.iterrows():
        print(f"   â€¢ {trabajo}: {datos_trab['Participacion_Mujeres']:.1f}% mujeres | {datos_trab['Total']:,.0f} empleados")

if len(feminizados) > 0:
    print(f"\nğŸŸ¢ SECTORES FEMINIZADOS (70-85% mujeres): {len(feminizados)}")
    for trabajo, datos_trab in feminizados.iterrows():
        print(f"   â€¢ {trabajo}: {datos_trab['Participacion_Mujeres']:.1f}% mujeres | {datos_trab['Total']:,.0f} empleados")

if len(feminizados_extremo) > 0:
    print(f"\nğŸŸ£ SECTORES ALTAMENTE FEMINIZADOS (â‰¥85% mujeres): {len(feminizados_extremo)}")
    for trabajo, datos_trab in feminizados_extremo.iterrows():
        print(f"   â€¢ {trabajo}: {datos_trab['Participacion_Mujeres']:.1f}% mujeres | {datos_trab['Total']:,.0f} empleados")

# EstadÃ­sticas ejecutivas clave
participacion_promedio = genero_pivot['Participacion_Mujeres'].mean()
mediana_participacion = genero_pivot['Participacion_Mujeres'].median()
sectores_equilibrados_count = len(equilibrados)

print("\nğŸ“Š INDICADORES CLAVE:")
print("-" * 30)
print(f"ğŸ“ˆ ParticipaciÃ³n promedio de mujeres: {participacion_promedio:.1f}%")
print(f"ğŸ“Š ParticipaciÃ³n mediana: {mediana_participacion:.1f}%")
print(f"âš–ï¸ Sectores equilibrados: {sectores_equilibrados_count} de {len(genero_pivot)} ({(sectores_equilibrados_count/len(genero_pivot)*100):.1f}%)")

# Sectores con mayor y menor participaciÃ³n femenina
sector_mas_femenino = genero_pivot.loc[genero_pivot['Participacion_Mujeres'].idxmax()]
sector_mas_masculino = genero_pivot.loc[genero_pivot['Participacion_Mujeres'].idxmin()]

print(f"\nğŸ† SECTOR MÃS FEMENINO: {sector_mas_femenino.name}")
print(f"   â€¢ {sector_mas_femenino['Participacion_Mujeres']:.1f}% mujeres | {sector_mas_femenino['Total']:,.0f} empleados")

print(f"\nğŸ”½ SECTOR MÃS MASCULINO: {sector_mas_masculino.name}")
print(f"   â€¢ {sector_mas_masculino['Participacion_Mujeres']:.1f}% mujeres | {sector_mas_masculino['Total']:,.0f} empleados")

print("\nğŸ’¡ INTERPRETACIÃ“N ESTRATÃ‰GICA:")
print("-" * 40)
if participacion_promedio < 40:
    interpretacion = "BAJA - Predominio masculino en el mercado laboral regional"
    recomendacion = "Implementar polÃ­ticas activas de inclusiÃ³n femenina"
elif participacion_promedio < 50:
    interpretacion = "MODERADA - Mercado laboral con sesgo masculino"
    recomendacion = "Fortalecer programas de equidad de gÃ©nero"
else:
    interpretacion = "EQUILIBRADA - ParticipaciÃ³n femenina adecuada"
    recomendacion = "Mantener y consolidar avances en equidad"

print(f"â€¢ EvaluaciÃ³n general: {interpretacion}")
print(f"â€¢ RecomendaciÃ³n estratÃ©gica: {recomendacion}")

print(f"\nğŸš€ RECOMENDACIONES EJECUTIVAS:")
print("-" * 40)
print("â€¢ Incrementar participaciÃ³n femenina en sectores masculinizados")
print("â€¢ Promover liderazgo femenino en sectores equilibrados")
print("â€¢ Desarrollar programas de capacitaciÃ³n en oficios tÃ©cnicos")
print("â€¢ Crear polÃ­ticas de conciliaciÃ³n trabajo-familia")
print("â€¢ Establecer metas de equidad de gÃ©nero por sector")

# â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬
# ğŸ­ ANÃLISIS SECTORIAL: ESTRUCTURA DEL EMPLEO REGIONAL - THE ECONOMIST
# â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬

print("ğŸ¯ ANÃLISIS SECTORIAL - ESTRUCTURA DEL EMPLEO")
print("="*60)
print("ğŸ“Š Procesando datos sectoriales con estÃ¡ndares The Economist...")

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ“Š PREPARACIÃ“N DE DATOS SECTORIALES
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

# Filtrar datos sectoriales (excluyendo total agregado)
datos_sectoriales = datos[
    (datos['Grupo_Simple'] != 'TOTAL OCUPADOS') & 
    (datos['Sexo'] == 'Ambos sexos')
].copy()

# AgregaciÃ³n por sector con validaciÃ³n de duplicados
empleo_por_sector = datos_sectoriales.groupby('Grupo_Simple')['Value'].sum().reset_index()
empleo_por_sector = empleo_por_sector.sort_values('Value', ascending=True)

# Eliminar duplicados y verificar integridad
empleo_por_sector = empleo_por_sector.drop_duplicates(subset=['Grupo_Simple'], keep='first').reset_index(drop=True)

print(f"âœ… Sectores procesados: {len(empleo_por_sector)}")
print(f"ğŸ“Š Empleo total sectorial: {empleo_por_sector['Value'].sum():,.0f}")

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ¨ CONFIGURACIÃ“N VISUAL AVANZADA
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

# Calcular mÃ©tricas de concentraciÃ³n
empleo_total = empleo_por_sector['Value'].sum()
empleo_por_sector['Porcentaje'] = (empleo_por_sector['Value'] / empleo_total) * 100
empleo_por_sector['Porcentaje_Acum'] = empleo_por_sector['Porcentaje'].cumsum()

# Paleta sectorial premium estilo The Economist
PALETA_THE_ECONOMIST_SECTORIAL = [
    '#1e3a8a',  # Azul marino ejecutivo
    '#7f1d1d',  # Burdeos elegante
    '#14532d',  # Verde bosque sofisticado
    '#d97706',  # Ãmbar dorado
    '#334155',  # Pizarra premium
    '#dc2626',  # CarmesÃ­ clÃ¡sico
    '#047857',  # Esmeralda corporativo
    '#ca8a04',  # Oro antiguo
    '#0f766e',  # Verde azulado elegante
    '#7c3aed'   # Violeta ejecutivo
]

# Asegurar que tenemos suficientes colores
n_sectores = len(empleo_por_sector)
if n_sectores > len(PALETA_THE_ECONOMIST_SECTORIAL):
    # Extender paleta si es necesario - ciclar colores
    colores_extendidos = (PALETA_THE_ECONOMIST_SECTORIAL * ((n_sectores // len(PALETA_THE_ECONOMIST_SECTORIAL)) + 1))
    colores_sectores = colores_extendidos[:n_sectores]
else:
    colores_sectores = PALETA_THE_ECONOMIST_SECTORIAL[:n_sectores]

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ·ï¸ OPTIMIZACIÃ“N DE ETIQUETAS
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

def optimizar_etiquetas_sectoriales(sectores, max_chars=22):
    """Crea etiquetas optimizadas para visualizaciÃ³n sectorial"""
    etiquetas_optimizadas = []
    
    for sector in sectores:
        if len(sector) <= max_chars:
            etiquetas_optimizadas.append(sector)
        elif ' y ' in sector:
            # Dividir en conectores naturales
            partes = sector.split(' y ')
            if len(partes) == 2 and all(len(p) <= max_chars for p in partes):
                etiquetas_optimizadas.append(f"{partes[0]}<br>y {partes[1]}")
            else:
                # Fallback: divisiÃ³n por palabras
                palabras = sector.split()
                mitad = len(palabras) // 2
                etiquetas_optimizadas.append(f"{' '.join(palabras[:mitad])}<br>{' '.join(palabras[mitad:])}")
        else:
            # DivisiÃ³n inteligente por palabras
            palabras = sector.split()
            if len(palabras) >= 3:
                mitad = len(palabras) // 2
                etiquetas_optimizadas.append(f"{' '.join(palabras[:mitad])}<br>{' '.join(palabras[mitad:])}")
            else:
                etiquetas_optimizadas.append(sector)
    
    return etiquetas_optimizadas

etiquetas_sectoriales = optimizar_etiquetas_sectoriales(empleo_por_sector['Grupo_Simple'])

print(f"ğŸ·ï¸ Etiquetas optimizadas: {len(etiquetas_sectoriales)}")

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ“Š CREACIÃ“N DE VISUALIZACIÃ“N EJECUTIVA
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

# Crear figura con configuraciÃ³n avanzada
fig_sectorial = go.Figure()

# Preparar datos para hover enriquecido
customdata_sectorial = np.column_stack([
    empleo_por_sector['Porcentaje'],
    empleo_por_sector['Porcentaje_Acum']
])

# GrÃ¡fico de barras horizontales premium estilo The Economist
fig_sectorial.add_trace(go.Bar(
    x=empleo_por_sector['Value'],
    y=etiquetas_sectoriales,
    orientation='h',
    marker=dict(
        color=colores_sectores,
        line=dict(color='white', width=1.2),
        opacity=0.85
    ),
    hovertemplate='<b>%{text}</b><br>' +
                  'Empleados: %{x:,.0f}<br>' +
                  'ParticipaciÃ³n: %{customdata[0]:.1f}%<br>' +
                  'Acumulado: %{customdata[1]:.1f}%<br>' +
                  '<extra></extra>',
    text=empleo_por_sector['Grupo_Simple'],
    customdata=customdata_sectorial,
    showlegend=False
))

# LÃ­nea de referencia para el promedio con estilo The Economist
promedio_sectorial = empleo_por_sector['Value'].mean()
fig_sectorial.add_vline(
    x=promedio_sectorial,
    line=dict(dash="dot", color='#64748b', width=2),
    annotation=dict(
        text=f"<b>Promedio<br>{promedio_sectorial:,.0f}</b>",
        textangle=0,
        font=dict(size=11, color='#1e293b', family='Georgia, serif'),
        bgcolor='rgba(255,255,255,0.95)',
        bordercolor='#64748b',
        borderwidth=1
    ),
    annotation_position="top"
)

# DiseÃ±o ejecutivo profesional estilo The Economist
fig_sectorial.update_layout(
    title=dict(
        text='<b>Estructura Sectorial del Empleo Regional</b><br>' +
             '<span style="font-size:14px; color:#64748b;">â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬</span><br>' +
             '<span style="font-size:13px; color:#475569;">DistribuciÃ³n de empleos por sector econÃ³mico â€¢ Los RÃ­os, Chile (2010-2019)</span>',
        font=dict(size=18, color='#1e293b', family='Georgia, serif'),
        x=0.12,
        xanchor='left',
        pad=dict(b=30)
    ),
    xaxis=dict(
        title=dict(
            text='<b>NÃºmero de Empleados</b>',
            font=dict(size=13, color='#334155', family='Georgia, serif')
        ),
        tickformat=',.0f',
        tickfont=dict(size=11, color='#475569', family='Georgia, serif'),
        gridcolor='#e2e8f0',
        gridwidth=1,
        showgrid=True,
        zeroline=False,
        linecolor='#cbd5e1',
        linewidth=1
    ),
    yaxis=dict(
        title='',
        tickfont=dict(size=10, color='#1e293b', family='Georgia, serif'),
        automargin=True,
        showgrid=False,
        categoryorder='array',
        categoryarray=etiquetas_sectoriales,
        linecolor='#cbd5e1',
        linewidth=1
    ),
    plot_bgcolor='#fefefe',
    paper_bgcolor='white',
    height=650,
    showlegend=False,
    margin=dict(l=280, r=140, t=160, b=90),
    font=dict(family='Georgia, serif', color='#1e293b')
)

fig_sectorial.show()

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ“ˆ ANÃLISIS EJECUTIVO AVANZADO
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

print("\n" + "="*75)
print("ğŸ“Š ANÃLISIS EJECUTIVO: ESTRUCTURA SECTORIAL DEL EMPLEO")
print("="*75)

# ClasificaciÃ³n estratÃ©gica de sectores
def clasificar_sector_estrategico(porcentaje):
    if porcentaje >= 20:
        return "ğŸ¯ CRÃTICO", "Sector de importancia sistÃ©mica"
    elif porcentaje >= 15:
        return "â­ ESTRATÃ‰GICO", "Sector de alta relevancia"
    elif porcentaje >= 10:
        return "ğŸ“ˆ IMPORTANTE", "Sector significativo"
    elif porcentaje >= 5:
        return "ğŸ”§ COMPLEMENTARIO", "Sector de apoyo"
    else:
        return "ğŸ“‹ ESPECIALIZADO", "Sector nicho"

print("\nğŸ† CLASIFICACIÃ“N ESTRATÃ‰GICA POR SECTOR:")
print("-" * 55)

sectores_criticos = 0
sectores_estrategicos = 0
sectores_importantes = 0

for _, sector in empleo_por_sector.iterrows():
    clasificacion, descripcion = clasificar_sector_estrategico(sector['Porcentaje'])
    
    # Conteo por categorÃ­a
    if "CRÃTICO" in clasificacion:
        sectores_criticos += 1
    elif "ESTRATÃ‰GICO" in clasificacion:
        sectores_estrategicos += 1
    elif "IMPORTANTE" in clasificacion:
        sectores_importantes += 1
    
    nombre_sector = sector['Grupo_Simple']
    if len(nombre_sector) > 35:
        nombre_sector = nombre_sector[:32] + "..."
    
    print(f"{clasificacion} {nombre_sector}")
    print(f"   â€¢ Empleados: {sector['Value']:,.0f} | ParticipaciÃ³n: {sector['Porcentaje']:.1f}%")
    print(f"   â€¢ {descripcion}")
    print()

# MÃ©tricas de concentraciÃ³n econÃ³mica
top_3_empleos = empleo_por_sector.tail(3)['Value'].sum()
top_5_empleos = empleo_por_sector.tail(5)['Value'].sum()

concentracion_top3 = (top_3_empleos / empleo_total) * 100
concentracion_top5 = (top_5_empleos / empleo_total) * 100

print("ğŸ“Š MÃ‰TRICAS DE CONCENTRACIÃ“N ECONÃ“MICA:")
print("-" * 45)
print(f"ğŸ“ˆ Total empleos analizados: {empleo_total:,.0f}")
print(f"ğŸ¯ ConcentraciÃ³n Top-3: {concentracion_top3:.1f}%")
print(f"ğŸ“Š ConcentraciÃ³n Top-5: {concentracion_top5:.1f}%")

# EvaluaciÃ³n de diversificaciÃ³n
if concentracion_top3 > 60:
    nivel_diversificacion = "BAJA - Riesgo de concentraciÃ³n alta"
    recomendacion_diversificacion = "ğŸš¨ Urgente: Diversificar base econÃ³mica"
elif concentracion_top3 > 50:
    nivel_diversificacion = "MODERADA - Vigilancia requerida"
    recomendacion_diversificacion = "âš ï¸ AtenciÃ³n: Promover sectores emergentes"
else:
    nivel_diversificacion = "ALTA - EconomÃ­a bien distribuida"
    recomendacion_diversificacion = "âœ… Estable: Mantener equilibrio actual"

print(f"\nğŸ’¡ EVALUACIÃ“N DE DIVERSIFICACIÃ“N:")
print("-" * 40)
print(f"â€¢ Nivel: {nivel_diversificacion}")
print(f"â€¢ RecomendaciÃ³n: {recomendacion_diversificacion}")

# Resumen ejecutivo por categorÃ­as
print(f"\nğŸ“‹ RESUMEN POR CATEGORÃAS ESTRATÃ‰GICAS:")
print("-" * 45)
print(f"ğŸ¯ Sectores crÃ­ticos: {sectores_criticos}")
print(f"â­ Sectores estratÃ©gicos: {sectores_estrategicos}")
print(f"ğŸ“ˆ Sectores importantes: {sectores_importantes}")
print(f"ğŸ”§ Otros sectores: {len(empleo_por_sector) - sectores_criticos - sectores_estrategicos - sectores_importantes}")

# Sectores dominantes
sector_dominante = empleo_por_sector.iloc[-1]
print(f"\nğŸ† SECTOR DOMINANTE: {sector_dominante['Grupo_Simple']}")
print(f"   â€¢ Empleados: {sector_dominante['Value']:,.0f}")
print(f"   â€¢ ParticipaciÃ³n: {sector_dominante['Porcentaje']:.1f}% del empleo regional")

print(f"\nğŸš€ RECOMENDACIONES ESTRATÃ‰GICAS:")
print("-" * 45)
print("â€¢ Monitorear salud de sectores crÃ­ticos regularmente")
print("â€¢ Desarrollar planes de contingencia para sectores dominantes")
print("â€¢ Promover innovaciÃ³n en sectores tradicionales")
print("â€¢ Identificar oportunidades de diversificaciÃ³n sectorial")
print("â€¢ Establecer indicadores de seguimiento por sector estratÃ©gico")

print(f"\nâœ… ANÃLISIS SECTORIAL COMPLETADO")
print("ğŸ¨ VisualizaciÃ³n estilo The Economist â€¢ Paleta sectorial diferenciada")
print("="*75)
````

````python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ‘¥ ANÃLISIS DE EQUIDAD DE GÃ‰NERO POR SECTOR LABORAL
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

print("ğŸ”„ Iniciando anÃ¡lisis de gÃ©nero...")

# Validar que los datos estÃ©n disponibles
if 'datos' not in locals():
    print("âš ï¸  Error: Los datos no estÃ¡n cargados. Ejecuta primero las celdas de carga de datos.")
else:
    print(f"âœ… Datos disponibles: {len(datos):,} registros")

# AnÃ¡lisis optimizado de gÃ©nero
try:
    # Filtrar datos de manera eficiente
    print("ğŸ“Š Procesando datos por gÃ©nero...")
    
    datos_genero = datos[
        (datos['Grupo_Simple'] != 'TOTAL OCUPADOS') & 
        (datos['Sexo'].isin(['Hombres', 'Mujeres']))
    ].copy()
    
    print(f"âœ… Datos filtrados: {len(datos_genero):,} registros")
    
    # Calcular totales de manera eficiente
    print("ğŸ”¢ Calculando totales por sector y gÃ©nero...")
    
    genero_trabajo = datos_genero.groupby(['Grupo_Simple', 'Sexo'])['Value'].sum().reset_index()
    genero_pivot = genero_trabajo.pivot(index='Grupo_Simple', columns='Sexo', values='Value').fillna(0)
    
    # Calcular participaciÃ³n femenina
    genero_pivot['Total'] = genero_pivot['Hombres'] + genero_pivot['Mujeres']
    genero_pivot['Participacion_Mujeres'] = (genero_pivot['Mujeres'] / genero_pivot['Total']) * 100
    
    # Filtrar trabajos con datos suficientes (umbral mÃ¡s bajo para mejor anÃ¡lisis)
    genero_pivot = genero_pivot[genero_pivot['Total'] > 100].copy()
    genero_pivot = genero_pivot.sort_values('Participacion_Mujeres', ascending=True)
    
    print(f"âœ… Sectores analizados: {len(genero_pivot)}")
    
    # Crear visualizaciÃ³n optimizada
    print("ğŸ“ˆ Creando visualizaciÃ³n...")
    
    # Asignar colores de manera eficiente
    def asignar_color_genero(participacion):
        if participacion < 25:
            return COLORES_GENERO['masculino_extremo']
        elif participacion < 40:
            return COLORES_GENERO['masculino_alto']
        elif participacion < 60:
            return COLORES_GENERO['equilibrado']
        else:
            return COLORES_GENERO['femenino_alto']
    
    colores_genero = [asignar_color_genero(p) for p in genero_pivot['Participacion_Mujeres']]
    
    # Crear grÃ¡fico optimizado
    fig = go.Figure()
    
    fig.add_trace(go.Bar(
        x=genero_pivot['Participacion_Mujeres'],
        y=genero_pivot.index,
        orientation='h',
        marker=dict(
            color=colores_genero,
            line=dict(color=COLORES_GENERO['equilibrado'], width=1),
            opacity=0.85
        ),
        hovertemplate='<b>%{y}</b><br>' +
                     'ParticipaciÃ³n Mujeres: %{x:.1f}%<br>' +
                     'Total Empleos: %{customdata:,.0f}<extra></extra>',
        customdata=genero_pivot['Total']
    ))
    
    # Layout optimizado estilo The Economist
    fig.update_layout(
        title=dict(
            text="<b>ParticipaciÃ³n de Mujeres por Sector Laboral</b><br>" +
                 "<span style='font-size:14px; color:#64748b;'>Los RÃ­os, Chile â€¢ AnÃ¡lisis de Equidad de GÃ©nero</span>",
            font=dict(size=16, color='#1e293b', family='Georgia, serif'),
            x=0.02,
            xanchor='left'
        ),
        xaxis=dict(
            title="<b>Porcentaje de Mujeres (%)</b>",
            tickfont=dict(size=10, color='#475569'),
            gridcolor='#e2e8f0',
            range=[0, 100]
        ),
        yaxis=dict(
            title="<b>Sector Laboral</b>",
            tickfont=dict(size=10, color='#1e293b'),
            automargin=True
        ),
        height=max(400, len(genero_pivot) * 30),
        showlegend=False,
        plot_bgcolor='#fefefe',
        paper_bgcolor='white',
        margin=dict(l=200, r=80, t=100, b=60)
    )
    
    fig.show()
    
    # AnÃ¡lisis estadÃ­stico optimizado
    print("\nğŸ“Š ANÃLISIS DE GÃ‰NERO POR SECTOR:")
    print("=" * 50)
    
    # Clasificar sectores de manera eficiente
    muy_masculino = genero_pivot[genero_pivot['Participacion_Mujeres'] < 25]
    masculino = genero_pivot[(genero_pivot['Participacion_Mujeres'] >= 25) & 
                           (genero_pivot['Participacion_Mujeres'] < 40)]
    equilibrado = genero_pivot[(genero_pivot['Participacion_Mujeres'] >= 40) & 
                             (genero_pivot['Participacion_Mujeres'] < 60)]
    femenino = genero_pivot[genero_pivot['Participacion_Mujeres'] >= 60]
    
    # Mostrar resultados de manera eficiente
    categorias = [
        ("ğŸ”´ Sectores muy masculinizados (<25% mujeres)", muy_masculino),
        ("ğŸŸ  Sectores masculinizados (25-40% mujeres)", masculino),
        ("ğŸŸ¡ Sectores equilibrados (40-60% mujeres)", equilibrado),
        ("ğŸŸ¢ Sectores feminizados (>60% mujeres)", femenino)
    ]
    
    for titulo, datos_cat in categorias:
        print(f"\n{titulo}: {len(datos_cat)}")
        for idx, fila in datos_cat.head(5).iterrows():  # Limitar a 5 para mejor rendimiento
            print(f"   â€¢ {idx}: {fila['Participacion_Mujeres']:.1f}%")
        if len(datos_cat) > 5:
            print(f"   ... y {len(datos_cat) - 5} sectores mÃ¡s")
    
    # EstadÃ­sticas resumen
    participacion_promedio = genero_pivot['Participacion_Mujeres'].mean()
    mediana = genero_pivot['Participacion_Mujeres'].median()
    
    print(f"\nğŸ“ˆ ESTADÃSTICAS RESUMEN:")
    print(f"   â€¢ ParticipaciÃ³n promedio de mujeres: {participacion_promedio:.1f}%")
    print(f"   â€¢ ParticipaciÃ³n mediana: {mediana:.1f}%")
    
    # Sectores extremos
    idx_equilibrado = (genero_pivot['Participacion_Mujeres'] - 50).abs().idxmin()
    sector_equilibrado = genero_pivot.loc[idx_equilibrado]
    
    print(f"   â€¢ Sector mÃ¡s equilibrado: {idx_equilibrado} ({sector_equilibrado['Participacion_Mujeres']:.1f}%)")
    print(f"   â€¢ Sector mÃ¡s masculinizado: {genero_pivot['Participacion_Mujeres'].min():.1f}%")
    print(f"   â€¢ Sector mÃ¡s feminizado: {genero_pivot['Participacion_Mujeres'].max():.1f}%")
    
    # Guardar variables globales para uso posterior
    globals()['trabajos_equilibrados'] = equilibrado
    globals()['trabajos_masculinos'] = masculino
    globals()['trabajos_femeninos'] = femenino
    
    print(f"\nâœ… AnÃ¡lisis completado exitosamente!")
    print(f"ğŸ“Š Variables creadas: trabajos_equilibrados, trabajos_masculinos, trabajos_femeninos")

except Exception as e:
    print(f"âŒ Error en el anÃ¡lisis: {str(e)}")
    print("ğŸ”§ Verifica que las celdas anteriores se hayan ejecutado correctamente")
````

````python
# ğŸ“ˆ ANÃLISIS TEMPORAL: EVOLUCIÃ“N SECTORIAL 2010-2019
# =====================================================

# â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬
# ğŸ“ˆ ANÃLISIS DE EVOLUCIÃ“N TEMPORAL - DISEÃ‘O THE ECONOMIST PREMIUM  
# â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬

print("ğŸ“ˆ ANÃLISIS DE EVOLUCIÃ“N TEMPORAL")
print("="*60)
print("ğŸ“Š Procesando series temporales con diseÃ±o The Economist...")

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ“Š PREPARACIÃ“N DE DATOS TEMPORALES
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

# Obtener los 6 sectores mÃ¡s importantes por volumen total
if 'empleo_por_sector' not in locals():
    # Recrear si no existe de anÃ¡lisis previo
    datos_sectoriales = datos[
        (datos['Grupo_Simple'] != 'TOTAL OCUPADOS') & 
        (datos['Sexo'] == 'Ambos sexos')
    ].copy()
    empleo_por_sector = datos_sectoriales.groupby('Grupo_Simple')['Value'].sum().reset_index()

top_6_sectores = empleo_por_sector.tail(6)['Grupo_Simple'].tolist()

print(f"ğŸ¯ Sectores analizados: {len(top_6_sectores)}")
for i, sector in enumerate(top_6_sectores, 1):
    print(f"   {i}. {sector}")

# Filtrar datos temporales para estos sectores
datos_temporales = datos[
    (datos['Grupo_Simple'].isin(top_6_sectores)) &
    (datos['Sexo'] == 'Ambos sexos')
].copy()

# AgregaciÃ³n por aÃ±o y sector
evolucion_sectorial = datos_temporales.groupby(['AÃ±o', 'Grupo_Simple'])['Value'].sum().reset_index()

print(f"âœ… Datos temporales procesados: {len(evolucion_sectorial)} registros")

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ¨ CONFIGURACIÃ“N VISUAL TEMPORAL AVANZADA
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

# Paleta temporal premium estilo The Economist
PALETA_TEMPORAL_THE_ECONOMIST = {
    'principal': '#1e3a8a',         # Azul navy ejecutivo
    'secundario': '#dc2626',        # Rojo clÃ¡sico The Economist  
    'terciario': '#047857',         # Verde esmeralda corporativo
    'cuaternario': '#d97706',       # Ãmbar dorado premium
    'quinario': '#7c3aed',          # Violeta ejecutivo
    'sexto': '#0f766e',            # Verde azulado elegante
    'septimo': '#be185d',          # Rosa ejecutivo
    'octavo': '#374151'            # Gris sofisticado
}

colores_temporales = list(PALETA_TEMPORAL_THE_ECONOMIST.values())[:len(top_6_sectores)]

# ConfiguraciÃ³n de estilos de lÃ­nea diferenciados
estilos_linea_temporal = [
    dict(width=3, dash='solid'),     # LÃ­nea sÃ³lida gruesa
    dict(width=3, dash='dash'),      # LÃ­nea discontinua
    dict(width=3, dash='dot'),       # LÃ­nea punteada
    dict(width=3, dash='dashdot'),   # LÃ­nea mixta
    dict(width=2.5, dash='solid'),   # LÃ­nea sÃ³lida media
    dict(width=2.5, dash='dash')     # LÃ­nea discontinua media
]

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ“Š CREACIÃ“N DE VISUALIZACIÃ“N TEMPORAL EJECUTIVA
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

fig_temporal = go.Figure()

# Datos para anÃ¡lisis de crecimiento
tasas_crecimiento = {}

# Crear series temporales para cada sector
for i, sector in enumerate(top_6_sectores):
    datos_sector = evolucion_sectorial[evolucion_sectorial['Grupo_Simple'] == sector]
    
    if len(datos_sector) >= 2:  # Validar datos suficientes
        # Calcular tasa de crecimiento
        valor_inicial = datos_sector['Value'].iloc[0]
        valor_final = datos_sector['Value'].iloc[-1]
        aÃ±os_transcurridos = len(datos_sector) - 1
        
        if valor_inicial > 0 and aÃ±os_transcurridos > 0:
            cagr = ((valor_final / valor_inicial) ** (1/aÃ±os_transcurridos) - 1) * 100
            tasas_crecimiento[sector] = cagr
        
        # Nombre optimizado para leyenda
        nombre_leyenda = sector if len(sector) <= 25 else sector[:22] + "..."
        
        # Agregar serie al grÃ¡fico
        fig_temporal.add_trace(go.Scatter(
            x=datos_sector['AÃ±o'],
            y=datos_sector['Value'],
            mode='lines+markers',
            name=nombre_leyenda,
            line=dict(
                color=colores_temporales[i],
                **estilos_linea_temporal[i % len(estilos_linea_temporal)]
            ),
            marker=dict(
                size=8,
                color=colores_temporales[i],
                opacity=0.8,
                line=dict(width=1, color='white')
            ),
            hovertemplate='<b>%{fullData.name}</b><br>' +
                         'AÃ±o: %{x}<br>' +
                         'Empleados: %{y:,.0f}<br>' +
                         '<extra></extra>',
            connectgaps=True
        ))

# Agregar lÃ­nea de referencia del promedio general
if evolucion_sectorial['Value'].mean() > 0:
    promedio_temporal = evolucion_sectorial.groupby('AÃ±o')['Value'].mean()
    
    fig_temporal.add_trace(go.Scatter(
        x=promedio_temporal.index,
        y=promedio_temporal.values,
        mode='lines',
        name='Promedio Regional',
        line=dict(
            color='rgba(122, 122, 122, 0.6)',
            width=2,
            dash='dot'
        ),
        hovertemplate='<b>Promedio Regional</b><br>' +
                     'AÃ±o: %{x}<br>' +
                     'Empleados: %{y:,.0f}<br>' +
                     '<extra></extra>',
        showlegend=True
    ))

# DiseÃ±o ejecutivo premium estilo The Economist
fig_temporal.update_layout(
    title=dict(
        text='<b>EvoluciÃ³n Temporal del Empleo por Sector</b><br>' +
             '<span style="font-size:14px; color:#64748b;">â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬â–¬</span><br>' +
             '<span style="font-size:13px; color:#475569;">DinÃ¡micas de crecimiento sectorial â€¢ Los RÃ­os, Chile (2010-2019)</span>',
        font=dict(size=18, color='#1e293b', family='Georgia, serif'),
        x=0.12,
        xanchor='left',
        pad=dict(b=30)
    ),
    xaxis=dict(
        title=dict(
            text='<b>AÃ±o</b>',
            font=dict(size=13, color='#334155', family='Georgia, serif')
        ),
        tickformat='d',
        tickfont=dict(size=11, color='#475569', family='Georgia, serif'),
        gridcolor='#e2e8f0',
        gridwidth=1,
        showgrid=True,
        dtick=1,  # Mostrar cada aÃ±o
        range=[2009.5, 2019.5],  # Margen visual
        linecolor='#cbd5e1',
        linewidth=1
    ),
    yaxis=dict(
        title=dict(
            text='<b>NÃºmero de Empleados</b>',
            font=dict(size=13, color='#334155', family='Georgia, serif')
        ),
        tickformat=',.0f',
        tickfont=dict(size=11, color='#475569', family='Georgia, serif'),
        gridcolor='#e2e8f0',
        gridwidth=1,
        showgrid=True,
        linecolor='#cbd5e1',
        linewidth=1
    ),
    plot_bgcolor='#fefefe',
    paper_bgcolor='white',
    height=580,
    showlegend=True,
    legend=dict(
        orientation="v",
        yanchor="top",
        y=0.98,
        xanchor="left",
        x=1.02,
        font=dict(size=10, color='#1e293b', family='Georgia, serif'),
        bgcolor='rgba(255,255,255,0.95)',
        bordercolor='#e2e8f0',
        borderwidth=1
    ),
    margin=dict(l=100, r=200, t=160, b=90),
    font=dict(family='Georgia, serif', color='#1e293b'),
    hovermode='x unified'
)

fig_temporal.show()

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ“ˆ ANÃLISIS TEMPORAL EJECUTIVO AVANZADO
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

print("\n" + "="*75)
print("ğŸ“ˆ ANÃLISIS EJECUTIVO: DINÃMICAS TEMPORALES SECTORIALES")
print("="*75)

print("\nğŸ¯ METODOLOGÃA TEMPORAL:")
print("â€¢ AnÃ¡lisis longitudinal de 6 sectores mÃ¡s relevantes")
print("â€¢ CÃ¡lculo de CAGR (Tasa de Crecimiento Anual Compuesta)")
print("â€¢ IdentificaciÃ³n de tendencias y puntos de inflexiÃ³n")
print("â€¢ ComparaciÃ³n con promedio regional")

# AnÃ¡lisis de tasas de crecimiento
if tasas_crecimiento:
    print("\nğŸ“Š TASAS DE CRECIMIENTO ANUAL (CAGR 2010-2019):")
    print("-" * 55)
    
    sectores_ordenados_crecimiento = sorted(tasas_crecimiento.items(), key=lambda x: x[1], reverse=True)
    
    sectores_en_crecimiento = []
    sectores_estables = []
    sectores_en_declive = []
    
    for sector, cagr in sectores_ordenados_crecimiento:
        if cagr > 10:
            tendencia = "ğŸ“ˆ CRECIMIENTO FUERTE"
            color_tendencia = "ğŸŸ¢"
            sectores_en_crecimiento.append(sector)
        elif cagr > 0:
            tendencia = "ğŸ“Š CRECIMIENTO MODERADO"
            color_tendencia = "ğŸ”µ"
            sectores_estables.append(sector)
        elif cagr > -5:
            tendencia = "â¡ï¸ ESTANCAMIENTO"
            color_tendencia = "ğŸŸ¡"
            sectores_estables.append(sector)
        elif cagr > -10:
            tendencia = "ğŸ“‰ DECLIVE MODERADO"
            color_tendencia = "ğŸŸ "
            sectores_en_declive.append(sector)
        else:
            tendencia = "â¬‡ï¸ DECLIVE FUERTE"
            color_tendencia = "ğŸ”´"
            sectores_en_declive.append(sector)
        
        nombre_sector = sector if len(sector) <= 40 else sector[:37] + "..."
        print(f"{color_tendencia} {tendencia}")
        print(f"   â€¢ {nombre_sector}")
        print(f"   â€¢ CAGR: {cagr:+.1f}% anual")
        print()
    
    # Resumen por categorÃ­as
    print("ğŸ“‹ RESUMEN DE TENDENCIAS:")
    print("-" * 30)
    print(f"ğŸ“ˆ Sectores en crecimiento: {len(sectores_en_crecimiento)}")
    print(f"â¡ï¸ Sectores estables: {len(sectores_estables)}")
    print(f"ğŸ“‰ Sectores en declive: {len(sectores_en_declive)}")
    
    # Sector con mayor y menor crecimiento
    if sectores_ordenados_crecimiento:
        mejor_sector, mejor_cagr = sectores_ordenados_crecimiento[0]
        peor_sector, peor_cagr = sectores_ordenados_crecimiento[-1]
        
        print(f"\nğŸ† SECTOR DE MAYOR CRECIMIENTO:")
        print(f"   â€¢ {mejor_sector}")
        print(f"   â€¢ CAGR: {mejor_cagr:+.1f}% anual")
        
        print(f"\nğŸ“‰ SECTOR DE MAYOR DECLIVE:")
        print(f"   â€¢ {peor_sector}")
        print(f"   â€¢ CAGR: {peor_cagr:+.1f}% anual")

# Recomendaciones estratÃ©gicas basadas en tendencias
print(f"\nğŸš€ RECOMENDACIONES ESTRATÃ‰GICAS:")
print("-" * 45)

if sectores_en_crecimiento:
    print("â€¢ Sectores en crecimiento:")
    print("  - Incrementar inversiÃ³n en formaciÃ³n especializada")
    print("  - Facilitar expansiÃ³n de empresas del sector")
    print("  - Atraer nuevas inversiones")

if sectores_estables:
    print("â€¢ Sectores estables:")
    print("  - Modernizar procesos y tecnologÃ­a")
    print("  - Diversificar oferta de servicios")
    print("  - Fortalecer competitividad")

if sectores_en_declive:
    print("â€¢ Sectores en declive:")
    print("  - Implementar programas de reconversiÃ³n laboral")
    print("  - Facilitar transiciÃ³n hacia sectores emergentes")
    print("  - Apoyo a innovaciÃ³n y transformaciÃ³n sectorial")

print(f"\nğŸ’¡ IMPLICACIONES PARA POLÃTICA REGIONAL:")
print("-" * 50)
print("â€¢ Diversificar hacia sectores de alto crecimiento")
print("â€¢ Establecer programas de reconversiÃ³n laboral")
print("â€¢ Monitorear indicadores tempranos de cambio sectorial")
print("â€¢ Promover innovaciÃ³n en sectores tradicionales")

print(f"\nâœ… ANÃLISIS TEMPORAL COMPLETADO")
print("ğŸ¨ VisualizaciÃ³n diferenciada por sector â€¢ Tendencias identificadas")
print("="*75)
````

> **ğŸ’¡ Insight Comparativo**: El anÃ¡lisis sectorial comparativo revela patrones especÃ­ficos de participaciÃ³n de gÃ©nero que requieren estrategias diferenciadas para maximizar el potencial del capital humano regional en cada sector estratÃ©gico.

````python
# ğŸ‘¥ ANÃLISIS COMPARATIVO DE GÃ‰NERO - SECTORES ESTRATÃ‰GICOS
# ==========================================================

print("ğŸ‘¥ ANÃLISIS COMPARATIVO DE GÃ‰NERO")
print("="*60)
print("ğŸ“Š Creando comparativo ejecutivo estilo The Economist...")

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ“Š PREPARACIÃ“N DE DATOS COMPARATIVOS
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

# Usar los top 6 sectores de anÃ¡lisis previo
if 'top_6_sectores' not in locals():
    # Recrear si no existe
    datos_sectoriales = datos[
        (datos['Grupo_Simple'] != 'TOTAL OCUPADOS') & 
        (datos['Sexo'] == 'Ambos sexos')
    ].copy()
    empleo_por_sector = datos_sectoriales.groupby('Grupo_Simple')['Value'].sum().reset_index()
    top_6_sectores = empleo_por_sector.tail(6)['Grupo_Simple'].tolist()

print(f"ğŸ¯ Sectores estratÃ©gicos analizados: {len(top_6_sectores)}")

# Filtrar datos por gÃ©nero para estos sectores
datos_genero_comparativo = datos[
    (datos['Grupo_Simple'].isin(top_6_sectores)) &
    (datos['Sexo'].isin(['Hombres', 'Mujeres']))
].copy()

# AgregaciÃ³n por sector y gÃ©nero
empleos_por_genero = datos_genero_comparativo.groupby(['Grupo_Simple', 'Sexo'])['Value'].sum().reset_index()

print(f"âœ… Datos comparativos procesados: {len(empleos_por_genero)} registros")

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ¨ CONFIGURACIÃ“N VISUAL COMPARATIVA AVANZADA
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

# Paleta de gÃ©nero ejecutiva diferenciada
PALETA_GENERO_COMPARATIVA = {
    'hombres': '#2E86AB',        # Azul corporativo profesional
    'mujeres': '#A23B72',        # Rosa ejecutivo elegante
    'referencia': '#7A7A7A'      # Gris neutro para referencias
}

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ·ï¸ OPTIMIZACIÃ“N DE ETIQUETAS COMPARATIVAS
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

def optimizar_etiquetas_comparativas(sectores, max_chars=20):
    """Optimiza etiquetas para visualizaciÃ³n comparativa"""
    etiquetas_comparativas = []
    
    for sector in sectores:
        if len(sector) <= max_chars:
            etiquetas_comparativas.append(sector)
        elif ' y ' in sector:
            partes = sector.split(' y ')
            if len(partes) == 2:
                etiquetas_comparativas.append(f"{partes[0]}<br>y {partes[1]}")
            else:
                palabras = sector.split()
                mitad = len(palabras) // 2
                etiquetas_comparativas.append(f"{' '.join(palabras[:mitad])}<br>{' '.join(palabras[mitad:])}")
        else:
            palabras = sector.split()
            if len(palabras) >= 3:
                mitad = len(palabras) // 2
                etiquetas_comparativas.append(f"{' '.join(palabras[:mitad])}<br>{' '.join(palabras[mitad:])}")
            else:
                etiquetas_comparativas.append(sector)
    
    return etiquetas_comparativas

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ“ˆ PREPARACIÃ“N DE DATOS ESTRUCTURADOS
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

# Estructurar datos por sector
sectores_estructurados = []
for sector in top_6_sectores:
    datos_sector = empleos_por_genero[empleos_por_genero['Grupo_Simple'] == sector]
    
    hombres_empleo = datos_sector[datos_sector['Sexo'] == 'Hombres']['Value'].sum()
    mujeres_empleo = datos_sector[datos_sector['Sexo'] == 'Mujeres']['Value'].sum()
    total_empleo = hombres_empleo + mujeres_empleo
    
    pct_hombres = (hombres_empleo / total_empleo * 100) if total_empleo > 0 else 0
    pct_mujeres = (mujeres_empleo / total_empleo * 100) if total_empleo > 0 else 0
    brecha_genero = abs(pct_hombres - pct_mujeres)
    
    sectores_estructurados.append({
        'sector': sector,
        'hombres': hombres_empleo,
        'mujeres': mujeres_empleo,
        'total': total_empleo,
        'pct_hombres': pct_hombres,
        'pct_mujeres': pct_mujeres,
        'brecha': brecha_genero
    })

# Ordenar por total de empleo para jerarquizaciÃ³n visual
sectores_estructurados = sorted(sectores_estructurados, key=lambda x: x['total'])

# Extraer datos para visualizaciÃ³n
etiquetas_comparativas = optimizar_etiquetas_comparativas([s['sector'] for s in sectores_estructurados])
valores_hombres = [s['hombres'] for s in sectores_estructurados]
valores_mujeres = [s['mujeres'] for s in sectores_estructurados]

print(f"ğŸ“Š Sectores estructurados: {len(sectores_estructurados)}")

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ“Š CREACIÃ“N DE VISUALIZACIÃ“N COMPARATIVA EJECUTIVA
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

fig_comparativo = go.Figure()

# Barra para hombres
fig_comparativo.add_trace(go.Bar(
    y=etiquetas_comparativas,
    x=valores_hombres,
    name='Hombres',
    orientation='h',
    marker=dict(
        color=PALETA_GENERO_COMPARATIVA['hombres'],
        line=dict(color='white', width=1.5),
        opacity=0.9
    ),
    hovertemplate='<b>ğŸ‘¨ Hombres</b><br>' +
                  'Sector: %{text}<br>' +
                  'Empleados: %{x:,.0f}<br>' +
                  'ParticipaciÃ³n: %{customdata:.1f}%<br>' +
                  '<extra></extra>',
    text=[s['sector'] for s in sectores_estructurados],
    customdata=[s['pct_hombres'] for s in sectores_estructurados]
))

# Barra para mujeres
fig_comparativo.add_trace(go.Bar(
    y=etiquetas_comparativas,
    x=valores_mujeres,
    name='Mujeres',
    orientation='h',
    marker=dict(
        color=PALETA_GENERO_COMPARATIVA['mujeres'],
        line=dict(color='white', width=1.5),
        opacity=0.9
    ),
    hovertemplate='<b>ğŸ‘© Mujeres</b><br>' +
                  'Sector: %{text}<br>' +
                  'Empleadas: %{x:,.0f}<br>' +
                  'ParticipaciÃ³n: %{customdata:.1f}%<br>' +
                  '<extra></extra>',
    text=[s['sector'] for s in sectores_estructurados],
    customdata=[s['pct_mujeres'] for s in sectores_estructurados]
))

# LÃ­nea de referencia del promedio general
promedio_general = sum([s['total'] for s in sectores_estructurados]) / len(sectores_estructurados) / 2

fig_comparativo.add_vline(
    x=promedio_general,
    line=dict(
        dash="dot", 
        color=PALETA_GENERO_COMPARATIVA['referencia'], 
        width=2
    ),
    annotation=dict(
        text=f"<b>Promedio<br>{promedio_general:,.0f}</b>",
        textangle=0,
        font=dict(size=11, color='#2c3e50'),
        bgcolor='rgba(255,255,255,0.9)',
        bordercolor=PALETA_GENERO_COMPARATIVA['referencia'],
        borderwidth=1
    ),
    annotation_position="top"
)

# DiseÃ±o ejecutivo profesional
fig_comparativo.update_layout(
    title=dict(
        text="â–¬â–¬â–¬â–¬ ParticipaciÃ³n Laboral por GÃ©nero en Sectores EstratÃ©gicos<br><sup style='color:#666666'>AnÃ¡lisis comparativo de empleo masculino vs femenino en Los RÃ­os</sup>",
        font=dict(family="Georgia", size=18, color="#2C3E50"),
        x=0.02,
        y=0.95
    ),
    xaxis=dict(
        title=dict(
            text='<b>NÃºmero de Personas Empleadas</b>',
            font=dict(size=14, color='#2c3e50')
        ),
        tickformat=',.0f',
        tickfont=dict(size=12, color='#34495e'),
        gridcolor='#ecf0f1',
        gridwidth=1,
        showgrid=True,
        zeroline=True,
        zerolinecolor='#bdc3c7',
        zerolinewidth=1
    ),
    yaxis=dict(
        title='',
        tickfont=dict(size=11, color='#2c3e50', family='Arial'),
        automargin=True,
        categoryorder='array',
        categoryarray=etiquetas_comparativas
    ),
    barmode='group',
    bargap=0.3,
    bargroupgap=0.15,
    plot_bgcolor='white',
    paper_bgcolor='white',
    height=650,
    showlegend=True,
    legend=dict(
        orientation="h",
        yanchor="bottom",
        y=1.02,
        xanchor="center",
        x=0.5,
        font=dict(size=13, color='#2c3e50'),
        bgcolor='rgba(255,255,255,0.95)',
        bordercolor='#ecf0f1',
        borderwidth=1
    ),
    margin=dict(l=200, r=100, t=140, b=80),
    font=dict(family='Georgia', color='#2c3e50'),
    annotations=[
        dict(
            text="Fuente: INE Chile | MetodologÃ­a: AnÃ¡lisis comparativo de gÃ©nero",
            showarrow=False,
            x=1.0, y=-0.12,
            xref="paper", yref="paper",
            xanchor="right",
            font=dict(size=10, color="#7F8C8D", family="Georgia", style="italic")
        )
    ]
)

fig_comparativo.show()

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ“ˆ ANÃLISIS COMPARATIVO EJECUTIVO AVANZADO
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

print("\n" + "="*75)
print("ğŸ‘¥ ANÃLISIS EJECUTIVO: PARTICIPACIÃ“N LABORAL POR GÃ‰NERO")
print("="*75)

print("\nğŸ“‹ METODOLOGÃA COMPARATIVA:")
print("â€¢ AnÃ¡lisis de los 6 sectores econÃ³micamente mÃ¡s relevantes")
print("â€¢ ComparaciÃ³n directa de volÃºmenes absolutos por gÃ©nero")
print("â€¢ CÃ¡lculo de brechas y participaciÃ³n relativa")
print("â€¢ ClasificaciÃ³n por nivel de equilibrio de gÃ©nero")

# AnÃ¡lisis de brechas por sector
print("\nğŸ“Š ANÃLISIS DE BRECHAS POR SECTOR:")
print("-" * 50)

sectores_equilibrados = 0
sectores_moderados = 0
sectores_sesgados = 0
sectores_extremos = 0

for sector_data in sectores_estructurados:
    brecha = sector_data['brecha']
    sector_nombre = sector_data['sector']
    
    if brecha < 10:
        categoria = "ğŸŸ¢ EQUILIBRADO"
        estrategia = "Mantener balance actual"
        sectores_equilibrados += 1
    elif brecha < 25:
        categoria = "ğŸŸ¡ MODERADAMENTE SESGADO"
        estrategia = "Programas de inclusiÃ³n graduales"
        sectores_moderados += 1
    elif brecha < 40:
        categoria = "ğŸŸ  ALTAMENTE SESGADO"
        estrategia = "Intervenciones especÃ­ficas urgentes"
        sectores_sesgados += 1
    else:
        categoria = "ğŸ”´ EXTREMADAMENTE SESGADO"
        estrategia = "PolÃ­ticas de transformaciÃ³n estructural"
        sectores_extremos += 1
    
    # Determinar gÃ©nero dominante
    if sector_data['pct_hombres'] > sector_data['pct_mujeres']:
        dominante = f"ğŸ‘¨ Masculino ({sector_data['pct_hombres']:.1f}%)"
    else:
        dominante = f"ğŸ‘© Femenino ({sector_data['pct_mujeres']:.1f}%)"
    
    nombre_mostrar = sector_nombre if len(sector_nombre) <= 35 else sector_nombre[:32] + "..."
    
    print(f"{categoria}")
    print(f"   â€¢ Sector: {nombre_mostrar}")
    print(f"   â€¢ Brecha de gÃ©nero: {brecha:.1f}%")
    print(f"   â€¢ GÃ©nero dominante: {dominante}")
    print(f"   â€¢ Total empleos: {sector_data['total']:,.0f}")
    print(f"   â€¢ Estrategia: {estrategia}")
    print()

# MÃ©tricas resumen
total_empleos_hombres = sum(valores_hombres)
total_empleos_mujeres = sum(valores_mujeres)
total_empleos_comparativo = total_empleos_hombres + total_empleos_mujeres

participacion_masc_total = (total_empleos_hombres / total_empleos_comparativo) * 100
participacion_fem_total = (total_empleos_mujeres / total_empleos_comparativo) * 100

print("ğŸ“Š MÃ‰TRICAS RESUMEN COMPARATIVO:")
print("-" * 40)
print(f"ğŸ‘¥ Total empleos analizados: {total_empleos_comparativo:,.0f}")
print(f"ğŸ‘¨ ParticipaciÃ³n masculina total: {participacion_masc_total:.1f}% ({total_empleos_hombres:,.0f})")
print(f"ğŸ‘© ParticipaciÃ³n femenina total: {participacion_fem_total:.1f}% ({total_empleos_mujeres:,.0f})")
print(f"âš–ï¸ Brecha general: {abs(participacion_masc_total - participacion_fem_total):.1f}%")

print(f"\nğŸ“ˆ DISTRIBUCIÃ“N POR CATEGORÃAS:")
print("-" * 35)
print(f"ğŸŸ¢ Sectores equilibrados: {sectores_equilibrados} ({sectores_equilibrados/len(sectores_estructurados)*100:.1f}%)")
print(f"ğŸŸ¡ Moderadamente sesgados: {sectores_moderados} ({sectores_moderados/len(sectores_estructurados)*100:.1f}%)")
print(f"ğŸŸ  Altamente sesgados: {sectores_sesgados} ({sectores_sesgados/len(sectores_estructurados)*100:.1f}%)")
print(f"ğŸ”´ Extremadamente sesgados: {sectores_extremos} ({sectores_extremos/len(sectores_estructurados)*100:.1f}%)")

# Sector mÃ¡s equilibrado y mÃ¡s desequilibrado
sector_mas_equilibrado = min(sectores_estructurados, key=lambda x: x['brecha'])
sector_mas_desequilibrado = max(sectores_estructurados, key=lambda x: x['brecha'])

print(f"\nğŸ† SECTOR MÃS EQUILIBRADO:")
print(f"   â€¢ {sector_mas_equilibrado['sector']}")
print(f"   â€¢ Brecha: {sector_mas_equilibrado['brecha']:.1f}%")
print(f"   â€¢ H: {sector_mas_equilibrado['pct_hombres']:.1f}% | M: {sector_mas_equilibrado['pct_mujeres']:.1f}%")

print(f"\nâš ï¸ SECTOR MÃS DESEQUILIBRADO:")
print(f"   â€¢ {sector_mas_desequilibrado['sector']}")
print(f"   â€¢ Brecha: {sector_mas_desequilibrado['brecha']:.1f}%")
print(f"   â€¢ H: {sector_mas_desequilibrado['pct_hombres']:.1f}% | M: {sector_mas_desequilibrado['pct_mujeres']:.1f}%")

print(f"\nğŸš€ RECOMENDACIONES EJECUTIVAS POR PRIORIDAD:")
print("-" * 55)
print("1. Intervenir urgentemente en sectores extremadamente sesgados")
print("2. Implementar programas graduales en sectores moderadamente sesgados")
print("3. Consolidar equilibrio en sectores ya balanceados")
print("4. Establecer metas especÃ­ficas de equidad por sector")
print("5. Monitorear indicadores de progreso trimestralmente")

print(f"\nâœ… ANÃLISIS COMPARATIVO COMPLETADO")
print("ğŸ¨ Paleta diferenciada por gÃ©nero â€¢ AnÃ¡lisis de brechas estratÃ©gicas")
print("="*75)
````
